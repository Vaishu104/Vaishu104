**Weather data classification**


downloaded the daily_weather dataset

Importing the libraries

```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

importing the file into the environment

```python
data = pd.read_csv('/content/daily_weather.csv')
```

storing the dataset

```python
data.head()
```

knowing the columns

```python
data.columns
```

chechking weather there is a null values in the dataset

```python
data[data.isnull().any(axis=1)].head()
```

removing unwanted column (Unnamed: 11) having full of empty values

```python
del data['Unnamed: 11']
```

classifying the columns

```python
data.columns
```

```python
data[data.isnull().any(axis=1)].head()
```

cleaning the number row which can not help us making any decision

```python
del data['number']
```

calculating the amount of data before removing the empty values (null values)

```python
before_rows = data.shape[0]
print(before_rows)
```

removing the null values from the data set

```python
data = data.dropna()
```

calculating the amount of data after the removal of null values

```python
after_rows = data.shape[0]
print(after_rows)
```

calculating the number of null values that have been presented in the dataset

```python
before_rows - after_rows
```

now we are going to filter the values if the value is below 25 the it is 0 and '1' if it is above 25.

```python
clean_data = data.copy()
clean_data['high_humidity_label'] = (clean_data['relative_humidity_3pm'] > 25)*1
clean_data['high_humidity_label'].head()
```

```python
y = clean_data[['high_humidity_label']].copy()
y.head()
```

```python
clean_data['relative_humidity_3pm'].head()
```

```python
y.head()
```

collecting all the morning data othe than humidity at 3pm

```python
morning_features = ['air_pressure_9am','air_temp_9am','avg_wind_direction_9am','avg_wind_speed_9am','max_wind_direction_9am','max_wind_speed_9am','rain_accumulation_9am','rain_duration_9am','relative_humidity_9am']
```

copying the dataset from y to x contains of morning data

```python
x=clean_data[morning_features].copy()
```

```python
x.columns
```

```python
y.columns
```

performing test and train split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=324)
```

test set

now we have made the machine to get decision using the classifier

```python
humidity_classifier = DecisionTreeClassifier(max_leaf_nodes=10,random_state=0)
humidity_classifier.fit(x_train,y_train)
```

```python
type(humidity_classifier)
```

now predicting on y test set

```python
y_predicted = humidity_classifier.predict(x_test)
```

the values for y predicted

```python
y_predicted[:10]
```

comparing it against the original y-test value

```python
y_test['high_humidity_label'][:10]
```

calculating the accuracy score of the model

```python
accuracy_score(y_test,y_predicted)*100
```

A classification model is built to predict the futuristic humidity level with present data and the accuracy of the model is noted.
